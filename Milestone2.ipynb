```python
# Install necessary packages

!pip install playwright nest_asyncio
!playwright install chromium
!apt-get install libatk1.0-0 libatk-bridge2.0-0 libatspi2.0-0 libxcomposite1

# Common imports and Colab event loop fix
import asyncio, json, csv, time
from pathlib import Path
import nest_asyncio
nest_asyncio.apply()
from playwright.async_api import async_playwright

#AJAX_URL = "https://webscraper.io/test-sites/e-commerce/ajax"  # target page
#AJAX_URL = "https://webscraper.io/test-sites"
AJAX_URL = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=1"
async def scrape_ajax_site():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)   # set False to watch it
        ctx = await browser.new_context()
        page = await ctx.new_page()
        await page.goto(AJAX_URL, timeout=60000)

        # Wait for initial product grid
        await page.wait_for_selector(".thumbnail", timeout=30000)

        # Keep clicking "Load more" until the button disappears or is disabled
        while True:
            try:
                # Button sometimes labeled "Load more" or has class .btn-load-more
                load_more = await page.query_selector("button:has-text('Load more')")
                if not load_more:
                    break
                is_disabled = await load_more.get_attribute("disabled")
                if is_disabled:
                    break

                # Click and wait for new items to appear
                before_count = len(await page.query_selector_all(".thumbnail"))
                await load_more.click()
                # Wait for network idle and more thumbnails to load
                await page.wait_for_load_state("networkidle")
                # Soft wait: poll until count increases or timeout
                for _ in range(30):
                    after_count = len(await page.query_selector_all(".thumbnail"))
                    if after_count > before_count:
                        break
                    await asyncio.sleep(0.2)
                # If no increase, assume end
                if after_count <= before_count:
                    break

            except Exception:
                # If the button isn't present or any error occurs, exit the loop
                break

        # Extract product cards
        cards = await page.query_selector_all(".thumbnail")
        rows = []

        for card in cards:
            # Title text and product link
            title_el = await card.query_selector(".title")
            title = (await title_el.text_content()).strip() if title_el else None
            url = await title_el.get_attribute("href") if title_el else None
            # Price
            price_el = await card.query_selector(".price")
            price = (await price_el.text_content()).strip() if price_el else None
            # Rating (count of .glyphicon-star)
            stars = await card.query_selector_all(".ratings .glyphicon-star")
            rating = len(stars) if stars else 0
            # Image
            img_el = await card.query_selector("img")
            img_src = await img_el.get_attribute("src") if img_el else None

            rows.append({
                "title": title,
                "price": price,
                "rating_stars": rating,
                "product_url": url,
                "image_url": img_src
            })

        await browser.close()
        return rows

# Run and save
data = asyncio.get_event_loop().run_until_complete(scrape_ajax_site())
print(f"Collected {len(data)} products")

# Save to CSV/JSON
Path("output").mkdir(exist_ok=True)
csv_path = Path("output/products_ajax.csv")
json_path = Path("output/products_ajax.json")

with open(csv_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)

with open(json_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"Saved CSV → {csv_path}")
print(f"Saved JSON → {json_path}")
```

```python
# Install necessary packages
!pip install playwright nest_asyncio
!playwright install chromium
!apt-get install libatk1.0-0 libatk-bridge2.0-0 libatspi2.0-0 libxcomposite1

# Common imports
import asyncio, csv
from pathlib import Path
import nest_asyncio
nest_asyncio.apply()
from playwright.async_api import async_playwright

# Base URL
STATIC_BASE_URL = "https://webscraper.io/test-sites/e-commerce/static"

async def scrape_fully_automatic_v2():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        all_products = []
        urls_to_scrape = []

        print("1. Visiting Home to Scan Categories...")
        await page.goto(STATIC_BASE_URL)

        # --- PHASE 1: AUTOMATIC DISCOVERY (CLICK & CHECK) ---
        # We find all top-level items in the sidebar
        # We must re-query elements inside the loop to avoid "stale element" errors
        # because clicking might refresh parts of the DOM.

        # First, just get the texts of the categories we want to check
        sidebar_texts = await page.eval_on_selector_all(
            '#side-menu > li > a',
            "elements => elements.map(e => e.innerText.trim())"
        )

        # Filter out "Home"
        categories_to_check = [t for t in sidebar_texts if t != "Home"]

        for cat_name in categories_to_check:
            print(f"   Checking Category: {cat_name}...")

            # Find the link again freshly
            # We use a specific locator for the link with this text
            category_link = page.locator(f"#side-menu > li > a:has-text('{cat_name}')")

            # 1. CLICK the category to ensure sub-menus expand
            # (We use try/except in case it's already open or unclickable)
            try:
                await category_link.click()
                await page.wait_for_timeout(500) # Wait for animation
            except Exception as e:
                print(f"     (Could not click {cat_name}, might be plain link)")

            # 2. CHECK for sub-links specifically under this active category
            # We look for the 'ul' sibling of the clicked link, or children in the active li
            # The structure usually becomes class="active" on the li

            # We look for any sub-links that are now visible under this category
            # (using xpath to find the UL following the A we just clicked)
            sub_links = await page.locator(f"//a[normalize-space(.)='{cat_name}']/following-sibling::ul//a").all()

            if len(sub_links) > 0:
                print(f"     -> Found {len(sub_links)} sub-categories. Adding them...")
                for sub in sub_links:
                    sub_name = (await sub.text_content()).strip()
                    sub_href = await sub.get_attribute("href")
                    full_url = f"https://webscraper.io{sub_href}"

                    urls_to_scrape.append({
                        "Main Category": cat_name,
                        "Sub Category": sub_name,
                        "URL": full_url
                    })
                    print(f"        -> Added: {sub_name}")
            else:
                # No sub-links found? Treat as a direct category
                href = await category_link.get_attribute("href")
                full_url = f"https://webscraper.io{href}"

                urls_to_scrape.append({
                    "Main Category": cat_name,
                    "Sub Category": "None",
                    "URL": full_url
                })
                print(f"     -> Added Direct Category: {cat_name}")

        # --- PHASE 2: SCRAPING LOOP ---
        print(f"\n2. Starting Scraping of {len(urls_to_scrape)} Categories...")

        for cat in urls_to_scrape:
            print(f"\n--- Processing: {cat['Sub Category']} ({cat['Main Category']}) ---")
            await page.goto(cat['URL'])

            # A. Dynamic Page Count
            pagination_links = await page.query_selector_all('.pagination li a')
            page_numbers = []
            for pl in pagination_links:
                txt = await pl.text_content()
                if txt.isdigit():
                    page_numbers.append(int(txt))

            total_pages = max(page_numbers) if page_numbers else 1
            print(f"   Total Pages detected: {total_pages}")

            # B. Page Loop
            for i in range(1, total_pages + 1):
                # Build URL
                joiner = "&" if "?" in cat['URL'] else "?"
                page_url = f"{cat['URL']}{joiner}page={i}"

                print(f"   Scraping Page {i}/{total_pages}: {page_url}")
                await page.goto(page_url)

                try:
                    await page.wait_for_selector(".thumbnail", timeout=3000)
                except:
                    print(f"   Skipping page {i} (No items found)")
                    continue

                # C. Extract Data
                cards = await page.query_selector_all(".thumbnail")
                for card in cards:
                    # Title
                    title_el = await card.query_selector(".title")
                    title = (await title_el.text_content()).strip() if title_el else "N/A"
                    # Price
                    price_el = await card.query_selector(".price")
                    price = (await price_el.text_content()).strip() if price_el else "N/A"
                    # Description
                    desc_el = await card.query_selector(".description")
                    desc = (await desc_el.text_content()).strip() if desc_el else "N/A"
                    # Rating
                    stars = await card.query_selector_all(".ratings .glyphicon-star")
                    rating = len(stars) if stars else 0

                    all_products.append({
                        "Main Category": cat['Main Category'],
                        "Sub Category": cat['Sub Category'],
                        "Title": title,
                        "Price": price,
                        "Description": desc,
                        "Rating": rating
                    })

        await browser.close()
        return all_products

# --- EXECUTION ---
print("Initializing Automatic Scraper V2...")
data = asyncio.get_event_loop().run_until_complete(scrape_fully_automatic_v2())
print(f"\nTask Complete. Collected {len(data)} total items.")

# --- PHASE 3: SAVING FILES ---
Path("output").mkdir(exist_ok=True)

# 1. Save EVERYTHING
csv_main = Path("output/final_task_products.csv")
if data:
    with open(csv_main, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)
    print(f"Saved Final CSV -> {csv_main}")

# 2. Save TABLETS ONLY (Separate File)
# We filter for any item where the Sub Category is "Tablets"
tablets_data = [item for item in data if "Tablets" in item['Sub Category']]
csv_tablets = Path("output/tablets_only.csv")

if tablets_data:
    with open(csv_tablets, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(tablets_data)
    print(f"Saved Tablets Data -> {csv_tablets} ({len(tablets_data)} items)")
else:
    print("No tablets found to save separately.")
```

```python
!pip install playwright nest_asyncio
!playwright install chromium
!apt-get install libatk1.0-0 libatk-bridge2.0-0 libatspi2.0-0 libxcomposite1

import asyncio
import csv
import re
from pathlib import Path
import nest_asyncio


nest_asyncio.apply()
from playwright.async_api import async_playwright

BASE_URL = "https://books.toscrape.com/"
MAX_BOOKS_TO_TEST = None

RATING_MAP = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

def calculate_dynamic_pricing(price_str, rating, category, availability_str):
    """
    Calculates discount based on Rating, Category, Stock, and Base Price.
    Returns: (New Price String, Discount %, Reason)
    """

    try:

        price = float(price_str.replace('£', ''))
    except:
        return price_str, 0, "Error parsing price"

    try:

        stock_match = re.search(r'(\d+)', availability_str)
        stock = int(stock_match.group(1)) if stock_match else 0
    except:
        stock = 0

    discount_pct = 0
    reasons = []

    # 2. Base Rating Rules
    if rating == 5:
        discount_pct = 20
        reasons.append("5-Star Special")
    elif rating == 4:
        discount_pct = 15
        reasons.append("Top Rated")

    # 3. Category Constraints (The "Children" rule)
    if category.lower() == "childrens":
        if discount_pct > 10:
            discount_pct = 10
            reasons.append("High Demand Category Cap")

    # 4. Inventory Protection (Scarcity)
    if stock < 5:
        discount_pct = 0
        reasons = ["Low Stock Alert (<5)"]

    # 5. Margin Protection (Low Price)
    if price < 15.00:
        discount_pct = 0
        reasons = ["Low Margin Item (<£15)"]

    # 6. Calculate Final Price
    if discount_pct > 0:
        discount_amount = (price * discount_pct) / 100
        new_price_val = price - discount_amount
        reason_str = " + ".join(reasons)
    else:
        new_price_val = price
        reason_str = "Standard Price" if not reasons else reasons[0]

    new_price_str = f"£{new_price_val:.2f}"

    return new_price_str, discount_pct, reason_str

async def get_product_details(browser, link):
    page = await browser.new_page()
    try:
        await page.goto(link)

        desc_el = await page.query_selector("#product_description + p")
        description = (await desc_el.text_content()).strip() if desc_el else "No Description"

        upc = "N/A"
        availability = "N/A"

        rows = await page.query_selector_all("table.table-striped tr")
        for row in rows:
            header = await row.query_selector("th")
            value = await row.query_selector("td")
            if not header or not value: continue

            header_text = (await header.text_content()).strip()
            val_text = (await value.text_content()).strip()

            if "UPC" in header_text:
                upc = val_text
            elif "Availability" in header_text:
                availability = val_text

        await page.close()
        return {"Description": description, "UPC": upc, "Availability": availability}

    except Exception:
        await page.close()
        return {"Description": "Error", "UPC": "Error", "Availability": "Error"}

async def scrape_books_logic():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        ctx = await browser.new_context(base_url=BASE_URL)
        page = await ctx.new_page()

        all_data = []
        books_collected = 0

        print("1. Visiting Home to Scan Categories...")
        await page.goto(BASE_URL)

        cat_elements = await page.query_selector_all('.side_categories > ul > li > ul > li > a')
        categories = []

        for el in cat_elements:
            name = (await el.text_content()).strip()
            href = await el.get_attribute("href")
            categories.append({"name": name, "url": BASE_URL + href})

        print(f"   -> Found {len(categories)} categories.")
        print(f"\n2. Starting Scraping & Pricing Analysis...\n")

        for cat in categories:
            if MAX_BOOKS_TO_TEST is not None and books_collected >= MAX_BOOKS_TO_TEST:
                break

            print(f"--- Processing: {cat['name']} ---")
            await page.goto(cat['url'])

            page_count = 1

            while True:
                if MAX_BOOKS_TO_TEST is not None and books_collected >= MAX_BOOKS_TO_TEST:
                    break

                product_pods = await page.query_selector_all(".product_pod")

                for pod in product_pods:
                    if MAX_BOOKS_TO_TEST is not None and books_collected >= MAX_BOOKS_TO_TEST:
                        break

                    title_el = await pod.query_selector("h3 a")
                    title = await title_el.get_attribute("title")

                    price_el = await pod.query_selector(".price_color")
                    price = (await price_el.text_content()).strip()

                    rating_el = await pod.query_selector(".star-rating")
                    rating_cls = await rating_el.get_attribute("class")
                    rating_word = rating_cls.replace("star-rating", "").strip()
                    rating_int = RATING_MAP.get(rating_word, 0)

                    link_relative = await title_el.get_attribute("href")
                    clean_link = link_relative.replace("../", "")
                    product_url = f"{BASE_URL}catalogue/{clean_link}"

                    details = await get_product_details(browser, product_url)

                    new_price, discount_pct, reason = calculate_dynamic_pricing(
                        price,
                        rating_int,
                        cat['name'],
                        details['Availability']
                    )

                    print(f"      [{books_collected + 1}] {title[:30]}... | Old: {price} -> New: {new_price} ({discount_pct}%)")

                    record = {
                        "Category": cat['name'],
                        "Title": title,
                        "Old Price": price,
                        "New Price": new_price,
                        "Discount (%)": f"{discount_pct}%",
                        "Pricing Logic": reason,
                        "Rating": rating_int,
                        "Availability": details['Availability'],
                        "UPC": details['UPC'],
                        "Description": details['Description'],
                        "URL": product_url
                    }
                    all_data.append(record)
                    books_collected += 1

                next_btn = await page.query_selector("li.next a")
                if next_btn:
                    await next_btn.click()
                    await page.wait_for_load_state("networkidle")
                    page_count += 1
                else:
                    break

        await browser.close()
        return all_data

limit_msg = "FULL CRAWL" if MAX_BOOKS_TO_TEST is None else f"LIMIT {MAX_BOOKS_TO_TEST}"
print(f"Initializing Scraper ({limit_msg})...")
data = asyncio.get_event_loop().run_until_complete(scrape_books_logic())
print(f"\nTask Complete. Collected {len(data)} total items.")

Path("output").mkdir(exist_ok=True)
csv_path = Path("output/smart_priced_books.csv")

if data:
    fieldnames = [
        "Category", "Title",
        "Old Price", "New Price", "Discount (%)", "Pricing Logic",
        "Rating", "Availability", "UPC", "Description", "URL"
    ]

    with open(csv_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)
    print(f"Saved CSV -> {csv_path}")
else:
    print("No data collected.")
```
